```python
import pandas as pd
import numpy as np

# describe the news dataset

df = pd.read_csv("../data/raw/AAPL_news_2024-06-01_2026-03-31.csv")

print(df.info())
print("\n...Info...")
print(df.tail())
print("...")
print(f"\nDuplicated rows found: {df.duplicated().sum()}")


df["eastern_dt"] = pd.to_datetime(df["eastern_dt"], utc=True).dt.tz_convert("US/Eastern")
df["date"] = df["eastern_dt"].dt.date

# calculate news articles by day
news_per_day= df.groupby("date").size()
all_business_days = pd.bdate_range(start=df["date"].min(), end=df["date"].max()).date
full_calendar_counts = news_per_day.reindex(all_business_days, fill_value=0)

distribution = full_calendar_counts.value_counts().sort_index()

news_counts = ',\n'.join([f"{v} days had {k} news article(s)" for k, v in distribution.items()])

print(f"News distribution:\n {news_counts}")
```
