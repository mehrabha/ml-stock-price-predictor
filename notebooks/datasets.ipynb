```python
import pandas as pd
import numpy as np

# describe the news dataset, fix time zones

df = pd.read_csv("../data/raw/apple_news_data.csv")

print(df.info())
print("\n...Info...")
print(df.head())
print("\n...Missing rows...")
print(df.isna().sum())
print("...")
print(f"\nDuplicated rows found: {df.duplicated().sum()}")


offsets = df["date"].str[-6:]
print(f"\n...timezone offset values to confirm utc timezomes...")
print(offsets.value_counts())

df["date"] = pd.to_datetime(df["date"], utc=True)
df["timestamp"] = df["date"].dt.tz_convert("US/Eastern")
df["date"] = df["timestamp"].dt.date

# calculate articles per day
print(f"\n...Number of times n articles appeared in a single day...")
articles_by_day = df.groupby("date").size().value_counts()
print(', '.join([f"{k}:{v}" for k, v in articles_by_day.items()]))
```
